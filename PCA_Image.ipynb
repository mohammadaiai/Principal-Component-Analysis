```python
# Import libraries
import matplotlib.pyplot as plt
import matplotlib.image as mplib
from sklearn.decomposition import PCA
import numpy as np
```

```python
# Read the image
img=mplib.imread("image.jpg") 
```

```python
# Print image shape
print("Original Shape:", img.shape)
```

```python
# Reshape the image to a 2-D matrix 
img_reshape=np.reshape(img, (img.shape[0], img.shape[1]*img.shape[2]))
```

```python
# Show the dimensions
print("Transformed Shape:", img_reshape.shape)
```

```python
# Train the PCA
pca=PCA(300).fit(img_reshape)
```

```python
# Transform the original image
img_transformed=pca.transform(img_reshape)
```

```python
# Show the transformed image shape 
print("Transformed Shape:", img_transformed.shape)
```

```python
# Use the inverse transform to reconstruct the original shape
img_inv=pca.inverse_transform(img_transformed)
```

```python
print("Transformed Shape:", img_inv.shape)
```

```python
# Craete the RGB image
img_final=np.reshape(img_inv, (img.shape[0], img.shape[1], img.shape[2]))
```

```python
# Show the final image
plt.axis("off");
plt.imshow(img_final.astype("uint8"))
plt.show
```
